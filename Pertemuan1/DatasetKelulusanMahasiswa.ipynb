```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cek colom data
print(data.columns)
```

```python
#Hapus spasi pada data
data.columns = data.columns.str.strip()
```

```python
# Korelasi antara Umur dan IPK
correlation_ipk = data['UMUR'].corr(data['IPK'])
print(f"Korelasi antara Umur dan IPK: {correlation_ipk}")
```

```python
# Scatter plot antara Umur dan IPK
plt.scatter(data['UMUR'], data['IPK'], alpha=0.5)
plt.title('Hubungan Umur dan IPK')
plt.xlabel('Umur')
plt.ylabel('IPK')
plt.show()
```

```python
# Korelasi antara Umur dan setiap IPS
correlation_ips1 = data['UMUR'].corr(data['IPS 1'])
correlation_ips2 = data['UMUR'].corr(data['IPS 2'])
correlation_ips3 = data['UMUR'].corr(data['IPS 3'])
correlation_ipk = data['UMUR'].corr(data['IPK'])

print(f"Korelasi antara Umur dan IPS 1: {correlation_ips1}")
print(f"Korelasi antara Umur dan IPS 2: {correlation_ips2}")
print(f"Korelasi antara Umur dan IPK: {correlation_ipk}")
```

```python
# Visual outlier

plt.figure(figsize=(12, 6)) #ngatur ukuran plot
sns.boxplot(data=data[['IPK']]) #buat box plot di kolom
plt.title('Box Plots') #judul
plt.show()
```
